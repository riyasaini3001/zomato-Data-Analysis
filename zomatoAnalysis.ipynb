```python
import re

# ---------------- WEEKDAY COLUMN ----------------
def extract_day_range(t):
    t = str(t)
    days = re.findall(r'(Mon|Tue|Wed|Thu|Fri|Sat|Sun)', t)

    if len(days) >= 2:
        return days[0] + " - " + days[-1]
    elif len(days) == 1:
        return days[0]
    else:
        return "Not Mentioned"

df['open_days'] = df['timings'].apply(extract_day_range)


# ---------------- TIME COLUMN ----------------
def extract_times(t):
    t = str(t).lower().replace(" ", "")

    # handle text
    t = t.replace("midnight","12:00am")
    t = t.replace("noon","12:00pm")

    # extract first two time values with optional minutes
    times = re.findall(r'(\d{1,2}(?::\d{2})?\s*[ap]m)', t)

    # Post-process to ensure minutes are always present (e.g., 8am -> 8:00am)
    processed_times = []
    for time_str in times:
        if re.match(r'^\d{1,2}\s*[ap]m$', time_str): # If minutes are missing
            # Replace 'am' with ':00am' and 'pm' with ':00pm'
            processed_times.append(time_str.replace('am', ':00am').replace('pm', ':00pm'))
        else:
            processed_times.append(time_str)

    if len(processed_times) >= 2:
        return processed_times[0], processed_times[1]
    return None, None

df['opening_time'], df['closing_time'] = zip(*df['timings'].apply(extract_times))

# convert to 12 hr proper format, handling errors gracefully
df['opening_time'] = pd.to_datetime(df['opening_time'], errors='coerce').dt.strftime('%I:%M %p')
df['closing_time'] = pd.to_datetime(df['closing_time'], errors='coerce').dt.strftime('%I:%M %p')

# Final single timing column
#df['open_timings'] = df['opening_time'] + " - " + df['closing_time']


# ---------------- SHOW RESULT ----------------
df[['timings','open_days']].head()
```
